## **1. Introduction**

Risk measurement in finance has traditionally been addressed within a classical statistical framework, with the Sharpe Ratio being one of the most widely used metrics. However, the limitations of the classical approach—which relies solely on variance—become evident in markets characterized by complex dynamics, variable correlations, and nonlinear structures.

This work proposes a generalization of the Sharpe Ratio through concepts from quantum mechanics and information theory, using the von Neumann entropy as a measure of structural risk. The proposal builds on a functional analogy between the correlation structure of assets and mixed quantum states, allowing risk assessment from a richer perspective with greater informational content.

<br><br><br>

## **2. Classical Sharpe Ratio Foundations**

The traditional Sharpe Ratio is given by:

<br>

$$
SR = \frac{\mathbb{E}[R - R_f]}{\sigma_R}
$$

<div align="right"> <strong>(1)</strong> </div>

Where:

- $R$ is the return of the asset or portfolio,  
- $R_f$ is the risk-free rate,  
- $\sigma_R$ is the standard deviation of returns.

This formulation assumes that risk is fully captured by variance, which holds under normal distributions but becomes inadequate in multimodal, fat-tailed, or dynamically correlated environments.

<br><br><br>

## **3. Limitations of the Classical Framework**

1. **Normality Assumption**: Variance fails to meaningfully reflect risk when returns are non-symmetric or multimodal.  
2. **Ignores Internal Correlations**: Portfolios of correlated assets may have low total variance but high structural risk.  
3. **Fails to Capture Dynamic Evolution**: Risk evolves based on the internal configuration of the system.

These limitations motivate the search for a metric that captures the **internal informational structure** of asset relationships.

<br><br><br>

In particular, the**density matrix formalism**, which extends statistical models of mixed states and provides a natural generalization of probabilistic descriptions with structural interdependence. ### **4.1 Quantum Formalism: Hilbert Spaces, Density Matrix, and Stochasticity**

- A **Hilbert space** $\mathcal{H}$ is a complex vector space with an inner product and completeness under its induced metric. It serves as the formal background for quantum systems and, by analogy, structured financial models.  

- The density matrix of a mixed state is defined as:

- The **density matrix** $\rho$ is a square $n \times n$ complex matrix acting on a Hilbert space $\mathcal{H}$, satisfying the properties:
  $$
  \rho = \rho^\dagger, \quad \rho \geq 0, \quad \mathrm{Tr}(\rho) = 1
  $$
  where $\rho^\dagger$ is the conjugate transpose of $\rho$.

- **Pure states** correspond to projectors onto single vectors $|\psi\rangle \in \mathcal{H}$, with:
  $$
  \rho = |\psi\rangle \langle \psi|, \quad \rho^2 = \rho, \quad \mathrm{Tr}(\rho^2) = 1, \quad S(\rho) = 0
  $$
  These states represent maximal knowledge (zero entropy) and no statistical mixture.

- **Mixed states** represent probabilistic mixtures of pure states:
  $$
  \rho = \sum_i p_i |\psi_i\rangle \langle \psi_i|, \quad p_i \geq 0, \quad \sum_i p_i = 1
  $$
  with
  $$
  \mathrm{Tr}(\rho^2) < 1, \quad S(\rho) = - \mathrm{Tr}(\rho \log \rho) > 0
  $$
  reflecting uncertainty or incomplete information about the system.

- **Superposition** refers to linear combinations of basis vectors in a pure state:
  $$
  |\psi\rangle = \sum_j \alpha_j |j\rangle, \quad \sum_j |\alpha_j|^2 = 1
  $$
  where $\{ |j\rangle \}$ is an orthonormal basis. Superposition encodes coherent coexistence of multiple configurations.

- **Entanglement** is a form of quantum correlation between subsystems $A$ and $B$, such that the joint state $\rho_{AB}$ cannot be factorized into:
  $$
  \rho_{AB} \neq \rho_A \otimes \rho_B
  $$
  where $\rho_A = \mathrm{Tr}_B(\rho_{AB})$ and $\rho_B = \mathrm{Tr}_A(\rho_{AB})$ are the reduced states. Entangled states display correlations stronger than any classical mixture, capturing deep structural dependencies essential in financial portfolio modeling.

- **Tensor Product Structure**: When modeling a portfolio composed of multiple subsystems (e.g., assets, strategies, or factors), their joint state space is given by the **tensor product** of individual Hilbert spaces:
  $$
  \mathcal{H}_{\text{total}} = \mathcal{H}_A \otimes \mathcal{H}_B
  $$
  A joint state $\rho_{AB}$ defined on $\mathcal{H}_{\text{total}}$ can exhibit both classical and quantum-like correlations depending on whether it factorizes as:
  $$
  \rho_{AB} = \rho_A \otimes \rho_B \quad \text{(uncorrelated)}
  $$
  or remains **entangled**:
  $$
  \rho_{AB} \neq \rho_A \otimes \rho_B
  $$
  This tensor product structure allows the encoding of **cross-asset dependencies** and **multi-agent interactions** beyond marginal statistics. Entropy computed over $\rho_{AB}$ then includes mutual information components, capturing **global interdependencies** within the portfolio architecture.


These quantum concepts provide a richer description of portfolio structure beyond classical statistics, enabling the Holistic Performance Ratio to measure risk as **structural complexity** rather than just dispersion.

<div align="right"> <strong>(3)</strong> </div>

It measures internal uncertainty: zero for pure states (maximum information) and maximal for maximally mixed states (structural ignorance).

## **5. Holistic Performance Ratio Definition**

We define the Holistic (Quantum) Sharpe Ratio as:

<br>

$$
HPR = \frac{\mathbb{E}[R - R_f]}{S(\rho)}
$$

<div align="right"> <strong>(4)</strong> </div>

Where $\rho$ represents the internal structure or latent embedding of the strategy.

#### **5.1 Interpretation**

- Numerator: average excess return.  
- Denominator: structural entropy as a proxy for complex risk.  

This ratio reflects **return per unit of structural disorder**, a generalization of volatility-based metrics.

Unlike traditional metrics that model risk as dispersion (e.g., volatility), HPR interprets risk as structural complexity—captured via entropy—which better reflects internal dependencies and dynamic adaptability of strategies.

<br><br><br>

## **6. Constructing the Density Matrix** $\rho$

There are multiple ways to construct $\rho$:

- **Empirical Correlation Matrices**: normalized to be positive semi-definite and with unit trace.  
- **Autoencoders or Deep Embeddings**: latent layers of trained models.  
- **Quantum Parametric States**: e.g., VQE or QAOA outputs.

The key requirement is that $\rho$ encodes **nontrivial internal dependencies** beyond mere variance.

<br>

### **6.1 Generalized Form: Matrix Representation and Constraints**

The density matrix $\rho$ is a Hermitian, positive semi-definite, trace-one operator defined as:

<br>

$$
\rho =
\begin{bmatrix}
\rho_{11} & \rho_{12} & \cdots & \rho_{1n} \\
\rho_{21} & \rho_{22} & \cdots & \rho_{2n} \\
\vdots    & \vdots    & \ddots & \vdots    \\
\rho_{n1} & \rho_{n2} & \cdots & \rho_{nn}
\end{bmatrix}
\in \mathbb{C}^{n \times n}
$$

<div align="right"><strong>(5)</strong></div>

With:

- **Hermiticity**: $\rho = \rho^\dagger$
- **Trace One**: $\mathrm{Tr}(\rho) = 1$  
- **Positive Semi-Definiteness**: all eigenvalues $\geq 0$

Alternatively, via spectral decomposition:

<br>

$$
\rho = \sum_{k=1}^m p_k |\psi_k\rangle \langle \psi_k|, \quad \sum_{k=1}^m p_k = 1
$$

<div align="right"><strong>(6)</strong></div>

Where each $|\psi_k\rangle$ in $\mathbb{C}^n$ is a normalized latent configuration. This compactly encodes **internal relational structure** (cross-strategy or cross-asset).

<br>

To turn a correlation matrix $C$ into a valid density matrix $\rho$, one can normalize it via


$$
\rho = \frac{C}{\mathrm{Tr}(C)}
$$


<div align="right"><strong>(7)</strong></div>

provided $C$ is positive semi-definite.

<br>

In multi-asset or multi-strategy portfolios, the global density matrix $\rho$ can be constructed via the tensor product of subsystems:

$$
\rho = \bigotimes_{i=1}^N \rho_i
$$

when the components are uncorrelated. If cross-dependencies exist, $\rho$ contains entanglement terms such that:

$$
\rho \neq \bigotimes_{i=1}^N \rho_i
$$

reflecting a non-separable internal structure across asset dimensions.

## **8. Functional Decomposition of $HPR$ : Dynamic-Informational Foundations**

The von Neumann entropy-based Sharpe Ratio can be expressed in functional form as:

<br>

$$
HPR [\rho(t)] = \frac{\int_0^T \mathbb{E}[R(t) - R_f] \, dt}{\int_0^T S(\rho(t)) \, dt}
$$

<div align="right"><strong>(8)</strong></div>

where $\rho(t)$ is a time-dependent density matrix.

<br>

#### **8.1 Functional Derivative of Risk: Quantum Entropy**

The temporal evolution of entropy is given by:

<br>

$$
\frac{dS(\rho)}{dt} = -\text{Tr}\left( \frac{d\rho}{dt} \log \rho + \rho \frac{d}{dt}(\log \rho) \right)
$$

<div align="right"><strong>(9)</strong></div>

Here $-k_B$ from Boltzman constant as local scaling can be adopted as a dimensionless scaling constant before Trace, inspired by statistical physics, to model entropy flow in analogy to thermodynamic dissipation—although no physical temperature is assumed, as addopting it would not meaningfully capture structural financial local dissipation.

**Financial interpretation:**

- Greater structural disorder implies higher entropy growth.  
- Robust strategies exhibit low rates of change in $\rho$.

<br>

#### **8.2 Derivative of** $HPR$: Sensitivity to Structure

The functional derivative of the ratio is:

<br>

$$
\frac{\delta HPR}{\delta \rho} = \frac{\partial \mathbb{E}[R]}{\partial \rho} \cdot \frac{1}{S(\rho)} - \mathbb{E}[R - R_f] \cdot \frac{1}{S(\rho)^2} \cdot \frac{\delta S(\rho)}{\delta \rho}
$$

<div align="right"><strong>(10)</strong></div>

This defines a **structural gradient of optimality**, useful in training quantum or hybrid models.

<br>

#### **8.3 Quantum-Classical Hybrid Systems**

The evolution of $\rho$ can follow a Lindblad-type equation:

<br>

$$
\frac{d\rho}{dt} = -i[\mathcal{H}, \rho] + \mathcal{D}(\rho)
$$

<div align="right"><strong>(11)</strong></div>

Where:

- $\mathcal{H}$: effective Hamiltonian (financial objective),  
- $\mathcal{D}(\rho)$: dissipative operator (noise, uncertainty).

The ratio $HPR$ can be interpreted as a quotient of **efficiency** (return) over **local dissipation** (entropy).

<br>

#### **8.4 Variational Formalism**

We define a Lagrangian-style action:

<br>

$$
\mathcal{L}[\rho(t)] = \int_0^T \left( \mathbb{E}[R(t) - R_f] - \lambda S(\rho(t)) \right) dt
$$

<div align="right"><strong>(12)</strong></div>

Its optimization satisfies:

<br>

$$
\frac{\delta \mathcal{L}}{\delta \rho(t)} = 0
$$

<div align="right"><strong>(13)</strong></div>

This connects theoretical physics principles to the training of structured financial strategies.

## **9. Removing Brownian Motion: A Global Formalism Without Energy Accumulation**

Classical financial models—from Bachelier to Black-Scholes—assume a local stochastic representation of risk via Brownian motion. Returns are modeled through:

<br>

$$
dX(t) = \mu\, dt + \sigma\, dW(t)
$$

<div align="right"><strong>(14)</strong></div>

where $(dW(t))$ is a Brownian increment with variance growing over time. However, this formulation lacks a **global perspective**: it describes local fluctuations without constraints on the system's total energy or information.

<br>

#### **9.1 Principle of Global Informational Stationarity**

Inspired by quantum mechanics, information geometry, and structural thermodynamics, we propose an alternative view:

In a holistic system, risk arises not from local randomness, but from the relational organization of its degrees of freedom.

Hence, total effective energy is preserved—not accumulated via Brownian diffusion, but dynamically redistributed under internal constraints:

<br>

$$
\delta \mathcal{E}_{\text{global}} = 0
$$

<div align="right"><strong>(15)</strong></div>

Where $(\mathcal{E}_{\text{global}})$ is an effective informational energy of the portfolio’s entire structure. Noise is replaced by a **self-consistent structural metric**.

<br>

#### **9.2 Risk Geometry and Informational Curvature**

From a geometric standpoint, Brownian motion corresponds to random paths on a flat manifold. In contrast, the holistic model posits that the system inhabits a **curved informational manifold**, with a metric induced by $(\rho)$.

The entropy $(S(\rho))$ acts as a volume measure (structural disorder) on that manifold. Thus, portfolio evolution follows geodesics in the space of mixed states—not random diffusion.

<br>

#### **9.3 Conserved Structure and Coherent Evolution**

Instead of assuming white noise and growing variance in time, we posit a coherent evolution of $(\rho)$:

<br>

$$
\frac{d\rho}{dt} = -i[H, \rho] + \mathcal{D}(\rho)
$$

<div align="right"><strong>(16)</strong></div>

Where:

- $[H, \rho]$ preserves total energy/information.  
- $\mathcal{D}(\rho)$ models **structured dissipation**, not arbitrary thermal noise.

The integrated dynamic entropy remains conserved under ideal conditions:

<br>

$$
\int_0^T \frac{dS(\rho(t))}{dt} \, dt = 0
$$

<div align="right"><strong>(17)</strong></div>

The system does not "accumulate noise" but rather reorganizes its internal complexity.

<br>

#### **9.4 Physical Analogy: From Brownian Motion to Unitary Evolution**

In quantum physics, a closed system evolves unitarily and does not generate entropy. Only open systems interacting with an environment exhibit entropy growth.

In our framework, the ideal portfolio behaves as a **closed, structurally conservative system**.

```python


# IMPORT LIBRARIES ---
import os
from alpaca.data.historical import StockHistoricalDataClient
from alpaca.data.requests import StockBarsRequest
from alpaca.data.timeframe import TimeFrame
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from numpy.linalg import eigvalsh

# --- ALPACA API CONFIGURATION ---
API_KEY = "PKIRUSO0K4IVVYPMLZHG"
SECRET_KEY = "PzKLj3j0dxLhDAQ24RQiSJJMQ7ahFQnRONHpjDaY"
client = StockHistoricalDataClient(API_KEY, SECRET_KEY)

# --- EXPERIMENT PARAMETERS ---
tickers = ["AAPL", "MSFT", "NVDA", "GOOGL", "AMZN"]
start_date = pd.Timestamp("2019-01-01", tz="America/New_York")
end_date = pd.Timestamp("2024-06-30", tz="America/New_York")

# --- DOWNLOAD DAILY CLOSING PRICES ---
request = StockBarsRequest(
    symbol_or_symbols=tickers,
    timeframe=TimeFrame.Day,
    start=start_date,
    end=end_date
)
bars_df = client.get_stock_bars(request).df
close_prices = bars_df['close'].unstack(level=0)

# --- 5. CALCULATE DAILY RETURNS ---
returns = close_prices.pct_change().dropna()

# --- 6. VWAP-BASED SIGNAL STRATEGY ---
vwap_linear = close_prices.expanding().mean()
signal = (close_prices > vwap_linear).astype(int)
signal = signal.replace(0, 1e-9)  # numerical stability

strategy_returns = (returns * signal.shift(1)).mean(axis=1).dropna()

# --- 7. VON NEUMANN ENTROPY FUNCTION ---
def von_neumann_entropy(rho_matrix):
    eigenvalues = eigvalsh(rho_matrix)
    eigenvalues = eigenvalues[eigenvalues > 1e-10]
    entropy = -np.sum(eigenvalues * np.log(eigenvalues))
    return entropy

# --- ROLLING WINDOW PARAMETERS ---
window = 60
risk_free_rate = 0.03 / 252
rolling_dates = strategy_returns.index[window:]

# Storage for metrics
hpr_series = []
sharpe_series = []
sp_series = []

# --- STORE DENSITY MATRICES ---
density_matrices = []
density_dates = []

# --- METRIC COMPUTATION LOOP ---
for i in range(window, len(strategy_returns)):
    window_returns = strategy_returns.iloc[i - window:i]
    window_asset_returns = (returns * signal.shift(1)).iloc[i - window:i].fillna(0)
    corr = window_asset_returns.corr().values
    rho = corr / np.trace(corr)

    # Store matrix and date
    density_matrices.append(rho)
    density_dates.append(strategy_returns.index[i])

    # Visualize last window matrix
    if i == len(strategy_returns) - 1:
        plt.figure(figsize=(7,6))
        sns.heatmap(pd.DataFrame(rho, index=tickers, columns=tickers), annot=True, cmap="coolwarm", fmt=".3f")
        plt.title("Density Matrix ρ (Normalized Correlation) - Last Rolling Window")
        plt.tight_layout()
        plt.show()

    S_rho = von_neumann_entropy(rho)
    mean_ret = window_returns.mean()
    excess_ret = mean_ret - risk_free_rate
    hpr = excess_ret / S_rho
    hpr_series.append(hpr)

    std_dev = window_returns.std()
    sharpe = excess_ret / std_dev
    sharpe_series.append(sharpe)

    current_cum_return = (1 + strategy_returns.iloc[:i]).prod()
    sp_series.append(current_cum_return)

# --- BASELINES: BUY & HOLD + RISK FREE ---
beta_returns = returns.mean(axis=1).loc[rolling_dates]
beta_cum_returns = (1 + beta_returns).cumprod()
risk_free_cum_returns = (1 + risk_free_rate) ** np.arange(len(rolling_dates))

# --- RESULTS DATAFRAME ---
metrics_df = pd.DataFrame({
    "HPR": hpr_series,
    "Sharpe": sharpe_series,
    "Strategy_CumReturn": sp_series,
    "Beta_CumReturn": beta_cum_returns.values,
    "RiskFree_CumReturn": risk_free_cum_returns
}, index=rolling_dates)

# --- APPROXIMATE STRUCTURAL ENTROPY FLOW ---
# Approximate entropy time derivative: ΔS/Δt
entropy_series = [von_neumann_entropy(rho) for rho in density_matrices]
entropy_derivative = np.gradient(entropy_series)

# Add it to the DataFrame
metrics_df["Entropy"] = entropy_series
metrics_df["Entropy_Derivative"] = entropy_derivative

metrics_df[["HPR", "Sharpe"]].to_csv("hpr_sharpe_comparison.csv")

# --- SAVE ALL DENSITY MATRICES IN FOLDER ---
output_folder = "density_matrices"
os.makedirs(output_folder, exist_ok=True)

# Save as .npy (3D array)
np.save(os.path.join(output_folder, "density_matrices.npy"), np.array(density_matrices))

# Optionally, save each matrix as individual CSV
for rho, date in zip(density_matrices, density_dates):
    df_rho = pd.DataFrame(rho, index=tickers, columns=tickers)
    date_str = date.strftime('%Y-%m-%d')
    df_rho.to_csv(os.path.join(output_folder, f"density_matrix_{date_str}.csv"))


# --- PLOT HPR VS SHARPE ---
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(metrics_df.index, metrics_df["HPR"], label="HPR", color="blue")
ax.plot(metrics_df.index, metrics_df["Sharpe"], label="Sharpe", color="green")
ax.set_title("HPR vs Sharpe (VWAP Linear Strategy)")
ax.set_xlabel("Date")
ax.set_ylabel("Value")
ax.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- PLOT CUMULATIVE RETURNS ---
plt.figure(figsize=(12, 6))
plt.plot(metrics_df.index, metrics_df["Strategy_CumReturn"], label="VWAP Linear Strategy", color="red")
plt.plot(metrics_df.index, metrics_df["Beta_CumReturn"], label="Beta (Buy & Hold)", color="blue")
plt.plot(metrics_df.index, metrics_df["RiskFree_CumReturn"], label="Risk-Free Asset", color="green")
plt.title("Cumulative Returns: Strategy vs Beta vs Risk-Free")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- GROWTH RATE COMPARISON ---
sharpe_growth = np.log(metrics_df["Sharpe"]).diff().dropna()
hpr_growth = np.log(metrics_df["HPR"].replace(0, np.nan)).diff().dropna()
common_index = sharpe_growth.index.intersection(hpr_growth.index)
sharpe_growth = sharpe_growth.loc[common_index]
hpr_growth = hpr_growth.loc[common_index]

# --- ROLLING CORRELATION OF GROWTH RATES ---
rolling_corr_window = 60
rolling_corr = sharpe_growth.rolling(window=rolling_corr_window).corr(hpr_growth)

# --- PLOT ROLLING CORRELATION ---
plt.figure(figsize=(12,6))
plt.plot(rolling_corr.index, rolling_corr, color='purple')
plt.title(f"Rolling Correlation ({rolling_corr_window} days) between Sharpe and HPR Growth")
plt.xlabel("Date")
plt.ylabel("Correlation")
plt.axhline(0, linestyle='--', color='black', alpha=0.7)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# --- PLOT ENTROPY DERIVATIVE ---
plt.figure(figsize=(12,6))
plt.plot(metrics_df.index, metrics_df["Entropy_Derivative"], label="dS(ρ)/dt", color="darkorange")
plt.axhline(0, linestyle='--', color='black', alpha=0.7)
plt.title("Structural Entropy Rate of Change (Approximated)")
plt.xlabel("Date")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
