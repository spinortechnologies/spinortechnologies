## **1. Motivation: Geometry in Financial Derivatives**

Markets do not evolve along uniform axes. A 1% move in 1 minute is not equivalent to 1% in 4 hours. Likewise, high-volume trades carry more weight than illiquid ones. Classical derivatives (based on fixed time) ignore:

- **Execution density (volume)**  
- **Signal volatility (pace of change)**  
- **Tick-based irregularity**  
- **Structural co-integration deviations**

In strategies like pairs trading or HFT arbitrage, a reversion signal may appear or disappear entirely depending on the frame in which it is measured. Traditional tools fail to detect these local slope dynamics because they lack **gradient invariance under reparametrization**.

> SMGC defines a gradient that is invariant to distortions in the domain — enabling arbitrage detection across heterogeneous regimes.

<br><br><br>

## **2. Rescaled Derivative Formalism**

Let $f(\vec{u})$ be a multivariate signal defined over $\vec{u} = (u_1, u_2, ..., u_n)$, where each $u_i$ is an axis (e.g., time, volume, volatility).

We define a **scaling map** $g_i: u_i \mapsto g_i(u_i)$ over each axis.

Then, the **scaled partial derivative** is:

<br>

$$
D^*_{u_i} f := \frac{\partial f}{\partial g_i(u_i)} = \frac{ \frac{\partial f}{\partial u_i} }{ \frac{dg_i}{du_i} }
$$

<div align="right"><strong>(1)</strong></div>

The full **scaled gradient** becomes:

<br>

$$
\nabla^* f = \left( D^*_{u_1} f, D^*_{u_2} f, ..., D^*_{u_n} f \right)
$$

<div align="right"><strong>(2)</strong></div>

This generalization preserves directional signal behavior under irregular structural evolution. It's particularly useful when signals are compared across timeframes, liquidity regimes, or volatility clusters.

<br><br><br>

## **3. Slope-Control Functional for Arbitrage Signals**

To extract arbitrage opportunities or enforce regularization across a co-integrated signal, we define the slope alignment functional:

<br>

$$
\mathcal{E}[f] = \int_{\Omega} \left\| \nabla^* f(\vec{u}) - \vec{S}(\vec{u}) \right\|^2 d\mu(\vec{u})
$$

<div align="right"><strong>(3)</strong></div>

Where:

- $\vec{S}(\vec{u})$ is a **target slope field**, constant or learned  
- $d\mu(\vec{u}) = \prod_i dg_i(u_i)$ is the **scaled differential volume**  
- $\Omega$ is a market region (e.g., session, execution window)  

This functional penalizes slope deviations, guiding signal optimization (or pricing behavior) toward desired flow geometries.

> It becomes a variational tool for slope-regularized signal reconstruction or portfolio rebalancing over distorted domains.

<br><br><br>

## **4. Reconstructing Normalized Signals**

To rebuild a consistent signal from its scaled gradient:

<br>

$$
f_{\text{norm}}(\vec{u}) = \int_{\mathcal{C}} \nabla^* f(\vec{u}) \cdot d\vec{g}(\vec{u})
$$

<div align="right"><strong>(4)</strong></div>

Where $\mathcal{C}$ is a path in the rescaled space.

This enables **signal comparison across assets**, normalized co-integration reversion tracking, and **transformation-invariant price flow estimation**.

<br><br><br>

## **5. Scaling Functions Without Distortion**

We propose **parametric linear scaling** to avoid information loss:

| Axis             | Variable             | Scaling Function $g_i(u_i)$                  | Purpose                                       |
|------------------|----------------------|----------------------------------------------|-----------------------------------------------|
| Time $t$         | Raw timestamps        | $g_t(t) = t$                                  | Identity: no distortion                       |
| Volume $v$       | Cumulative contracts  | $g_v(v) = v$                                  | Execution-aware normalization                 |
| Volatility $\sigma$ | Std deviation     | $g_\sigma(t) = \int_0^t \sigma(u)\,du$        | Spreads time in high-volatility periods       |
| Ticks $n$        | Event index           | $g_n(n) = n$                                  | Event-based uniformization                    |

> These mappings enable a **coordinate system aligned with market microstructure**, critical for real-time arbitrage logic.

<br><br><br>

## **6. Divergence and Field Behavior**

The divergence of a scaled gradient field quantifies **expansion vs compression zones**:

<br>

$$
\text{div}(\nabla^* f) = \sum_i \frac{\partial D^*_{u_i} f}{\partial u_i}
$$

<div align="right"><strong>(5)</strong></div>

This operator can expose:

- Execution pressure zones  
- Mean-reversion anomalies  
- Slippage compression regions  
- Arbitrageable dislocations in spread evolution

For instance, in a spread $s(t) = p_1 - \beta p_2$, computing $\text{div}(\nabla^* s)$ over $(t,v,\sigma)$ reveals structural cointegration violations.

<br><br><br>

## **7. Application: Scaled Gradient Cointegration (SGC)**

Let $s(t)$ be the spread of two co-integrated assets. Its raw derivative may fail under regime shifts. Applying scaled derivatives:

<br>

$$
\nabla^* s = \left( D^*_t s, D^*_v s, D^*_\sigma s \right)
$$

<div align="right"><strong>(6)</strong></div>

We define a **scaled mispricing intensity functional**:

<br>

$$
\mathcal{M}(t) := \left\| \nabla^* s(t) \right\|^2
$$

<div align="right"><strong>(7)</strong></div>

This reveals when slope-based decouplings emerge, serving as entry signals in stat-arb or mean-reverting portfolios.

> Traditional co-integration only uses statistical stationarity; SGC uses **geometry-aware differential consistency**.

<br><br><br>

## **8. Summary of Operators**

- **Scaled Derivative:**
  $$
  D^*_{u_i} f = \frac{\partial f}{\partial g_i(u_i)}
  $$

- **Scaled Gradient:**
  $$
  \nabla^* f = \left( D^*_{u_1} f, ..., D^*_{u_n} f \right)
  $$

- **Slope Alignment Functional:**
  $$
  \mathcal{E}[f] = \int \| \nabla^* f - \vec{S} \|^2 d\mu
  $$

- **Signal Reconstruction:**
  $$
  f_{\text{norm}} = \int \nabla^* f \cdot d\vec{g}
  $$

- **Mispricing Intensity Functional:**
  $$
  \mathcal{M}(t) = \left\| \nabla^* s(t) \right\|^2
  $$

- **Divergence Operator:**
  $$
  \text{div}(\nabla^* f) = \sum_i \frac{\partial D^*_{u_i} f}{\partial u_i}
  $$

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from alpaca.data.historical import StockHistoricalDataClient
from alpaca.data.requests import StockBarsRequest
from alpaca.data.timeframe import TimeFrame


# --- ALPACA API CONFIGURATION ---
API_KEY = "PKKXIWB03HMYVSIHEYY3"
SECRET_KEY = "xu6dMaD16ITuAqUFu3eYuMnceezf5UgeZuUFh8Lp"
client = StockHistoricalDataClient(API_KEY, SECRET_KEY)


# --- PARAMETERS ---
ticker = "NVDA"
start_date = pd.Timestamp("2022-01-01", tz="America/New_York")
end_date = pd.Timestamp("2024-06-30", tz="America/New_York")


# --- FETCH PRICE AND VOLUME DATA ---
request = StockBarsRequest(
    symbol_or_symbols=ticker,
    timeframe=TimeFrame.Day,
    start=start_date,
    end=end_date
)
bars_df = client.get_stock_bars(request).df

# --- PREPARE DATA ---
df = bars_df.xs(ticker, level=0)
df = df[['close', 'volume']].copy()
df['returns'] = df['close'].pct_change()
df['volatility'] = df['returns'].rolling(window=10).std()

# Drop NaNs
df.dropna(inplace=True)

# --- COMPUTE SCALING AXES ---
df['time_axis'] = (df.index.astype(np.int64) // 1e9).astype(float)  # seconds
df['cum_volume'] = df['volume'].cumsum()
df['int_volatility'] = df['volatility'].cumsum()

# --- APPLY SMGC DERIVATIVES ---
f = df['close'].values
u_t = df['time_axis'].values
u_v = df['cum_volume'].values
u_sigma = df['int_volatility'].values

df['D*_time'] = np.gradient(f, u_t)
df['D*_volume'] = np.gradient(f, u_v)
df['D*_volatility'] = np.gradient(f, u_sigma)

# --- PLOT SMGC GRADIENTS ---
plt.figure(figsize=(14, 8))

plt.subplot(3, 1, 1)
plt.plot(df.index, df['D*_time'], label="D*_time")
plt.title("Scaled Derivative w.r.t. Time")
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df.index, df['D*_volume'], color='green', label="D*_volume")
plt.title("Scaled Derivative w.r.t. Cumulative Volume")
plt.grid(True)
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df.index, df['D*_volatility'], color='red', label="D*_volatility")
plt.title("Scaled Derivative w.r.t. Integrated Volatility")
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Second-order derivatives for local 3D divergence estimation
df['∂_D*_time/∂t'] = np.gradient(df['D*_time'].values, df['time_axis'].values)
df['∂_D*_volume/∂v'] = np.gradient(df['D*_volume'].values, df['cum_volume'].values)
df['∂_D*_volatility/∂σ'] = np.gradient(df['D*_volatility'].values, df['int_volatility'].values)

# Compute SMGC 3D divergence
df['divergence'] = (
    df['∂_D*_time/∂t'] +
    df['∂_D*_volume/∂v'] +
    df['∂_D*_volatility/∂σ']
)

# Visualization
plt.figure(figsize=(12, 4))
plt.plot(df.index, df['divergence'], label='Divergence (∇·∇*f)', color='purple')
plt.axhline(0, linestyle='--', color='gray')
plt.title("SMGC Divergence — Expansion vs. Compression Zones")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
